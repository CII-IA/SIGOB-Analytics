```python
import polars as pl
import geopandas as gpd
from shapely.geometry import Point
```

```python
df = pl.read_parquet('data/iieg_data.parquet')
df = df.insert_column(0, pl.arange(0, df.height).alias("id"))
```

```python
df.filter(pl.col("x").is_not_null() & pl.col("y").is_not_null())
```

```python
colonias = gpd.read_file("data/SHAPE_jalisco_colonias/conjunto_de_datos/14as.shp")
print("santiago totolimixpan" in colonias["NOM_ASEN"].str.lower().unique())
colonias[colonias["NOM_ASEN"].str.lower() == "santiago totolimixpan"]
```

```python
colonias2 = gpd.read_file("data/SHAPEColonias20202024/ISDC2020_2024.shp")
print("santiago totolimixpan" in colonias2["NOMCOL1"].str.lower().unique())
colonias2[colonias2["NOMCOL1"].str.lower() == "santiago totolimixpan"]
```

```python
colonias_wgs84 = colonias.to_crs(epsg=4326)
p = Point(-102.926676, 20.390521)
polygon = colonias_wgs84.loc[colonias_wgs84["NOM_ASEN"].str.lower() == "santiago totolimixpan"].geometry.values[0]
polygon.contains(p)
```

```python
colonias2_wgs84 = colonias2.to_crs(epsg=4326)
p = Point(-102.926676, 20.390521)
polygon = colonias2_wgs84.loc[colonias2_wgs84["NOMCOL1"].str.lower() == "santiago totolimixpan"].geometry.values[0]
polygon.contains(p)
```
